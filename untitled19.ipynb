```python
%%writefile simd_branch.c
#include<stdio.h>
#include<stdlib.h>
#include<omp.h>
#define N 100000000
int main(){
    double*a=malloc(N*sizeof(double));
    double*b=malloc(N*sizeof(double));
    double*c=malloc(N*sizeof(double));

    for (int i = 0; i < N; i++){
      a[i]=(double)i/N;
      b[i]=(double)(N-i)/N;
    }
    double start=omp_get_wtime();

    for (int i = 0; i < N; i++){
      if (a[i]>0.5)
          c[i]=a[i]*b[i];
      else
          c[i]=a[i]+b[i];
}

double end=omp_get_wtime();
printf("Time = %f\n",end-start);
}
```

```python
%%writefile simd_branch.c
#include<stdio.h>
#include<stdlib.h>
#include<omp.h>
#define N 100000000
int main(){
    double*a=malloc(N*sizeof(double));
    double*b=malloc(N*sizeof(double));
    double*c=malloc(N*sizeof(double));

    for (int i = 0; i < N; i++){
      a[i]=(double)i/N;
      b[i]=(double)(N-i)/N;
    }
    double start=omp_get_wtime();
#pragma omp simd
for (int i = 0; i < N; i++){
  if (a[i] >0.5)
     (c[i] = a[i]*b[i]);
  else
     c[i] = a[i] = b[i];
}
double end = omp_get_wtime();
printf("Time = %f\n",end-start);
}
```

```python
import numpy as np
import time
```

```python
N = 100000000
A = np.random.rand(N)
B = np.zeros(N)
C = np.zeros(N)
```

```python
start = time.time()
for i in range(N):
  if A[i] > 0.5:
    B[i] = A[i] * 2
  else:
    B[i] = A[i] + 5
end = time.time()
print("time with branch",end - start)
```

```python
start = time.time()
B = np.where(A > 0.5, A * 2, A + 5)
end = time.time()
print("Vectorized",end - start)
```

```python
start=time.time()
for i in range(N):
  if A[i]>0.9:
    for j in range(50):
      B[i]+=j
  else:
    B[i]+=1
print("Imbalanced loop time",time.time()-start)
```

```python
start = time.time()
for i in range(N):
  C[i] = A[i] + B[i]
end = time.time()
print("Vectorized",end - start)
```

```python
for i in range(N):
  C[i]=A[i]+B[i]
end=time.time()
print("normal loop time:",end-start)
```

```python
start=time.time()
C=A+B
end=time.time()
print("vectorized (SIMD)  time:",end-start)
```
